```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ORIGINAL
```

```python
img1=np.zeros((300,500),dtype='uint8')
font=cv2.FONT_ITALIC=3
img2=cv2.putText(img1,"Kumaran",(5,70),font,3,(255),5,cv2.LINE_AA)
cv2.imshow("Original",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# EROSION
```

```python
kernel1=np.ones((5,5),np.uint8)
erode=cv2.erode(img2,kernel1)
cv2.imshow("Erosion1",erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel2=cv2.getStructuringElement(cv2.MORPH_CROSS,(7,7))
erode=cv2.erode(img2,kernel2)
cv2.imshow("Erosion2",erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# DILATION
```

```python
dilute=cv2.dilate(img2,kernel2)
cv2.imshow("Dilution",dilute)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
